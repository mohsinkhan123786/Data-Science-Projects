```
st1 = 'We will find all the "gmail" accounts and will ignore all other email account. "abc@gmail.com" abc@mailg.com xyz@ymail.com xyz@hotmail.com <another@gmail.com> the END'
pat = '[\w]+@gmail.com'
# a-z + A-Z + _
print(re.findall(pat, st1))

# print(re.search(pat, st))
# print(re.match(pat,st))
# re.split(pat, st)
```

```
str2 = '101 COM  not    Computer 201 ART Artist 401 MAT     Mathematics   501 PHY    Physics'
pt2 = '([0-9]{3})\s*([A-Z]{3})\s*([a-zA-Z]{4,})'

rslt = re.findall(pt2, str2)

c_code = input('Please enter the Course Code you want to search : ').upper()
srch = []
for i in rslt:
  if c_code in  i:
    srch.append(i)
print(srch)
```

```
mydata = requests.get('https://www.whatmobile.com.pk/')
mydata.text
```

```
#<a href="HTC_Mobiles_Prices" title="HTC mobile phones" class="lnk">HTC</a>
mobile='href="([a-zA-Z]+)'
x = re.findall(mobile, mydata.text)

mobile_companies = []
for i in x:
  if i not in mobile_companies:
    mobile_companies.append(i)

print(mobile_companies)
```

```
# </a>\n</h4>\n<span class="PriceFont">Rs. 43,999</span>\n</li>\n<li class="product">


prices = '"PriceFont">Rs. ([0-9,]+)'
price_list = re.findall(prices, mydata.text)
print(price_list)
print(min(price_list))
print(max(price_list)) # Why we get 69,999 as max value and not 248,000. One thing is because these are the string values stored in a list
```
